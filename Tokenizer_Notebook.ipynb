# Data Into CSV File

In this section I take the folder of .v files we have collected and then run just a script to convert it to text an store it into a csv

Using this csv I create a data frame we can use for tokenization.

```python
import os
import csv
import pandas as pd

from google.colab import drive
drive.mount('/content/drive')

PROJECT_ROOT_DRIVE = "/content/drive/MyDrive/Capstone/verilog"
CSV_PATH = "/content/drive/MyDrive/Capstone/data.csv"
```

Here we take each .v file and convert it to text format and then put it into a CSV by filename and by text.

We can use this later to tokenize each word to being next token prediction model

We can also start adding models here to test it and more tokenization code etc...

```python
def convert_v_files_to_csv(folder_path, output_csv_path):
  '''
  This converts .v to files to text and compiles them into one big
  csv file for tokenizing later
  '''

  v_files = [f for f in os.listdir(folder_path) if f.endswith('.v')]

  with open(output_csv_path, 'w', newline='') as csv_file:
            csv_writer = csv.writer(csv_file)

            csv_writer.writerow(['file name', 'text'])

            for v_file in v_files:
                v_file_path = os.path.join(folder_path, v_file)

                with open(v_file_path, 'r') as file_content:
                    content = file_content.read()

                csv_writer.writerow([v_file, content])
  print(f"Conversion to CSV successful: {output_csv_path}")

convert_v_files_to_csv(PROJECT_ROOT_DRIVE, CSV_PATH)
```

Putting this into a pd.DataFrame for future use

Note we can also do a split up of our .v files here into training and testing set but for now I'll keep it in one dataframe
